```python
class PlannerAgent:
    def create_schedule(self, tasks):
        return sorted(tasks, key=lambda x: x["priority"])


class ReminderAgent:
    def remind(self, tasks):
        reminders = []
        for task in tasks:
            reminders.append(
                f"Reminder: Complete '{task['task']}' before {task['deadline']}"
            )
        return reminders


class MotivationAgent:
    def motivate(self):
        return "You are capable of achieving your goals. Keep going!"


tasks = [
    {"task": "Complete Kaggle Assignment", "priority": 1, "deadline": "Tomorrow"},
    {"task": "Study Full Stack Development", "priority": 2, "deadline": "Tonight"},
    {"task": "Revise Notes", "priority": 3, "deadline": "Weekend"}
]

planner = PlannerAgent()
reminder = ReminderAgent()
motivator = MotivationAgent()

print("Study Schedule:")
for task in planner.create_schedule(tasks):
    print(task)

print("\nReminders:")
for message in reminder.remind(tasks):
    print(message)

print("\nMotivation:")
print(motivator.motivate())
```

```python
class SecurityAgent:
    def protect_data(self, user_data):
        protected_data = {}

        for key, value in user_data.items():
            if key.lower() in ["password", "email", "phone"]:
                protected_data[key] = "Hidden"
            else:
                protected_data[key] = value

        return protected_data


user_data = {
    "name": "Vaishnavi",
    "email": "example@gmail.com",
    "phone": "9876543210",
    "study_goal": "Become a Full Stack Developer"
}

security = SecurityAgent()

print("Protected User Data:")
print(security.protect_data(user_data))
```
